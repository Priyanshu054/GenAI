## Website Summerizer Using Ollama Llama3.2 Model
This notebook demonstrates how to use the Ollama Llama3.2 model to summarize the content of a website. The model processes the text extracted from the website and generates a concise summary.

```python
from openai import OpenAI
from website_scraper import Website
from IPython.display import Markdown, display
```

```python
!ollama serve
```

```python
# check if ollama server is running
import requests
requests.get("http://localhost:11434").content
```

```python
ollama = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
```

```python
system_prompt = "You are an assistant that analyzes the contents of a website \
and provides a short summary, ignoring text that might be navigation related. \
Respond in markdown."
```

```python
def user_prompt_for(website: Website) -> str:
    user_prompt = f"You are looking at a website titled {website.title}"
    user_prompt += "\nThe contents of this website is as follows; please provide a short summary of this website in markdown. If it includes news or announcements, then summarize these too.\n\n"
    user_prompt += website.text
    return user_prompt
```

```python
def messages_for(website: Website) -> list[dict]:
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_for(website)}
    ]
```

```python
# Function to summarize a website given its URL
def summarize(url: str, ollama_model) -> str:
    website = Website(url)
    response = ollama.chat.completions.create(
        model=ollama_model,
        messages=messages_for(website)
    )
    return response.choices[0].message.content
```

```python
def display_summary(url: str, ollama_model="llama3.2"):
    summary = summarize(url, ollama_model)
    display(Markdown(summary))
```

```python
url = "https://www.edwarddonner.com"

display_summary(url)
```

```python
url = "https://en.wikipedia.org/wiki/Artificial_intelligence"

display_summary(url)
```

```python
url = "https://en.wikipedia.org/wiki/Artificial_intelligence"

display_summary(url, ollama_model="gemma3")
```
